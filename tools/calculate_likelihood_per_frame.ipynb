## Load MFCC

```python
def load_mfcc(file):
    with open(file, 'r') as f:
        lines = f.readlines()
    mfcc = [re.sub(r'\n|\]', '', i).strip().split(' ') for i in lines[1:]]    
    return np.array(mfcc).astype(np.float)
    
mfcc = load_mfcc(lib_mfcc)
mfcc
```
